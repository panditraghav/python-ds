```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7]
y = [34,56,23,36,8,12, 78]
```

```python
plt.plot(x, y, color='red', marker='D', linestyle='--', markersize=20)
```
